```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL of the website to scrape
url = "https://jiji.ng/cars"
```

```python
# Send a GET request to the URL
response = requests.get(url)

#check status code to see if it connects, it must be 200
response
```

```python
#text format of response
response.text
```

```python
#
response.content
```

```python
 # Parse the HTML content of the page(it will change to html format)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Print the parsed HTML content
print(soup)
```

```python
import json 
import xmltojson
```

```python
# Get the page through get() method 
#html_response = requests.get(url=url, headers = headers) 
  
# Save the page content as sample.html:this will save it on the local drive
with open("jiji.html", "w", encoding="utf-8") as html_file: 
    html_file.write(response.text)
      
```

```python
with open("jiji.html", "r", encoding='utf-8',errors='ignore') as html_file: 
    html = html_file.read() 
    jiji_json = xmltojson.parse(html) 
      
```

```python
with open("data.json", "w") as file: 
    json.dump(jiji_json, file) 
      
print(jiji_json)
```
